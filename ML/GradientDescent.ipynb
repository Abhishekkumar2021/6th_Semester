# Gradient Descent

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
```

```python
def f(x):
    return x**2 - 7*x + 10

def df(x):
    return 2*x - 7

# This code implements gradient descent to find the minimum of a function.
# We assume that the function is differentiable and that we have access to its derivative.
# It takes three parameters: x0, the initial value of x; learning_rate, the learning rate; and n_iter, the number of iterations to perform.
# It returns an array of x values.

def gradient_descent(x0, learning_rate, n_iter):
    x = np.zeros(n_iter) # create an array to store x values
    x[0] = x0 # set the first value to x0
    for i in range(1, n_iter): # iterate from 1 to n_iter
        x[i] = x[i-1] - learning_rate * df(x[i-1]) # apply the gradient descent equation
    return x

x = np.linspace(-100, 100, 1000)
y = f(x)

x0 = 100
learning_rate = 0.9
n_iter = 15

x_min = gradient_descent(x0, learning_rate, n_iter)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b')

# plot a line between all the points in x_min
plt.plot(x_min, f(x_min), 'ro-') # why ro-? => red, circle, line
plt.show()
```
